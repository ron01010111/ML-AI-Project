```python
import pandas as pd
import matplotlib.pyplot as plt


Movie_data = pd.read_csv('MovieFranchises.csv')
```

```python
#Displays the studio with the highest average rating
avg_rating_by_studio = Movie_data.groupby('Studio')['VoteAvg'].mean()
avg_rating_by_studio.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Studio')
plt.xlabel('Studio')
plt.ylabel('Average Rating')
plt.show()

# Display the studio with the highest total vote count
total_vote_count_by_studio = Movie_data.groupby('Studio')['VoteCount'].sum()
total_vote_count_by_studio.plot(kind='bar' , color='lightgreen')
plt.title('Total Vote Count by Studio')
plt.xlabel('Studio')
plt.ylabel('Total Vote Count')
plt.show()
```

```python
# Clean and convert 'Lifetime Gross' column to numeric
Movie_data['Lifetime Gross'] = pd.to_numeric(Movie_data['Lifetime Gross'])

# Find the movie with the highest 'Lifetime Gross'
highest_grossing_movie = Movie_data.loc[Movie_data['Lifetime Gross'].idxmax()]

# Display a horizontal bar plot for the highest-grossing movie
plt.figure(figsize=(10, 6))
plt.barh(highest_grossing_movie['Title'], highest_grossing_movie['Lifetime Gross'], color='skyblue')
plt.title('Highest Grossing Movie')
plt.xlabel('Lifetime Gross(In Millions)')
plt.ylabel('Movie Title')
plt.show()

```

```python
# Creates a plot to observe the trend over time(compares liftime gross and the year it was released)
Movie_data['ReleaseDate'] = pd.to_datetime(Movie_data['ReleaseDate'], format='%m-%d-%y', errors='coerce')
plt.figure(figsize=(12, 6))
plt.plot(Movie_data['ReleaseDate'], Movie_data['Lifetime Gross'], marker='o', linestyle='-')
plt.title('Time Series: Lifetime Gross Over Time')
plt.xlabel('Release Date')
plt.ylabel('Lifetime Gross(In Millions)')
plt.show()


```

```python
# Studio Market Share based on Lifetime Gross
studio_market_share = Movie_data.groupby('Studio')['Lifetime Gross'].sum()
plt.figure(figsize=(8, 8))
plt.pie(studio_market_share, labels=studio_market_share.index, autopct='%1.1f%%', startangle=90)
plt.title('Studio Market Share based on Lifetime Gross')
plt.show()

```
